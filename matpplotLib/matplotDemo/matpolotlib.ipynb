```python
# Data
x = np.array(["A", "B", "C", "D", "E"])  # Categories
y = np.array([10, 20, 15, 25, 30])  # Values

colors = [10, 20, 30, 40, 50]  # Must match x and y length
sizes = [500, 100, 500, 100, 500]  # Must match x and y length

# Scatter plot
plt.scatter(x, y, c=colors, s=sizes, cmap="viridis")  # Fixed 'c' and colormap name
plt.colorbar()  # Adds color scale

# Labels and title
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Category")
plt.ylabel("Value")

plt.show()
```

```python
# Data
x = np.array(["A", "B", "C", "D", "E"])  # Categories
y = np.array([10, 20, 15, 25, 30])  # Values
z=np.array([20, 12, 17,20,15])  # Values2 array

colors = [10, 20, 30, 40, 50]  # Must match x and y length
sizes = [500, 100, 500, 100, 500]  # Must match x and y length

# Scatter plot
plt.scatter(x, y, c=colors, s=sizes, cmap="viridis")  # Fixed 'c' and colormap name

plt.scatter(x, z, color="r", s=sizes, cmap="viridis")  # Fixed 'c' and colormap name

t=plt.colorbar()  # Adds color scale
t.set_label("ColorBar",fontsize=15)

# Labels and title
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Category")
plt.ylabel("Value")

plt.show()
```
