```python
# # this is box plot 
# f= sns.load_dataset(" ")
# sns.boxplot(x=" ",y=" ",data=f)
# plt.show()
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",order=["Sun","Sat","Thur","Fri"])
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",
            color="r",order=["Sun","Sat","Thur","Fri"],showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",
            color="w",order=["Sun","Sat","Thur","Fri"],showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,
            color="w",order=["Sun","Sat","Thur","Fri"],showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="total_bill",y="day",data=var,
            color="w",order=["Sun","Sat","Thur","Fri"],showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma"
           ,orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(data=var,showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma"
           ,orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(y=var["total_bill"])
plt.show()
```

# styling plot  in seborn 

```python
# seaborn figure style
# removing axes spines 
# scale and context
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.boxplot(y=var["total_bill"])
plt.show()
```

```python
sns.set_style("darkgrid")# whitegrid
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
sns.set_style("white")# whitegrid
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
sns.despine()
plt.show()
```

```python
sns.set_style("white")# whitegrid
plt.figure(figsize=(12,3)) #3,12
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()

sns.despine()
plt.show()
```

```python
sns.set_style("white")# whitegrid
sns.set_context("poster",font_scale=0.2)
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()

sns.despine()
plt.show()
```

```python
sns.set_style("white")# whitegrid
sns.set_context("paper",font_scale=1)
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()

sns.despine()
plt.show()
```

```python
sns.set_style("white")# whitegrid
sns.set_context("paper",font_scale=1)
sns.barplot(x="day",y="total_bill",data=var,palette="cool")
plt.grid()

sns.despine()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day",palette="summer")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day",palette="summer")
fg.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day",palette="summer")
fg.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend()
plt.show()
```
